### 5. Analiza przekroczeń normy wg województw

Dodatkowo, sprawdzimy jak wygląda liczba dni z przekroczeniem normy PM2.5 w podziale na województwa. W tym celu mapujemy stacje pomiarowe do odpowiednich województw na podstawie metadanych. Wykres przedstawia średnią liczbę dni z przekroczeniem normy dla stacji w danym województwie.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Sprawdź czy plik z danymi (wynik Zadania 1) istnieje
csv_file = 'data_2015_2018_2021_2024.csv'

if not os.path.exists(csv_file):
    print(f"Plik {csv_file} nie istnieje. Proszę wykonać najpierw Zadanie 1 (funkcja prepare_common_data).")
else:
    try:
        # Wczytanie danych - zakładamy MultiIndex w kolumnach (Kod stacji, Miejscowość)
        df_all = pd.read_csv(csv_file, index_col=0, header=[0, 1], parse_dates=True)
        
        # Wczytanie metadanych
        meta = pd.read_excel('metadane_new.xlsx')
        
        # Znajdowanie nazw kolumn z Kodem stacji i Województwem
        col_kod = next((c for c in meta.columns if 'Kod stacji' in str(c)), 'Kod stacji')
        col_woj = next((c for c in meta.columns if 'Województwo' in str(c)), 'Województwo')
        
        # Mapowanie
        mapping = dict(zip(meta[col_kod], meta[col_woj]))
        
        # Obliczenia przekroczeń (> 15 ug/m3)
        res = {}
        years = sorted(df_all.index.year.unique())
        
        for y in years:
            sub = df_all[df_all.index.year == y]
            # Średnie dobowe
            daily = sub.groupby(sub.index.date).mean()
            # Liczba dni z przekroczeniem
            count = (daily > 15).sum()
            res[y] = count
            
        df_res = pd.DataFrame(res)
        # Index df_res to MultiIndex (Kod stacji, Miejscowość)
        
        # Dodanie kolumny Województwo
        df_res['Województwo'] = df_res.index.get_level_values(0).map(mapping)
        
        # Grupowanie po województwach (średnia dla stacji w województwie)
        df_woj = df_res.groupby('Województwo').mean()
        
        if df_woj.empty:
            print("Brak danych po zmapowaniu województw. Sprawdź poprawność mapowania.")
        else:
            # Wizualizacja
            df_long = df_woj.reset_index().melt(id_vars='Województwo', var_name='Rok', value_name='Dni')
            
            plt.figure(figsize=(14, 7))
            sns.barplot(data=df_long, x='Województwo', y='Dni', hue='Rok', palette='viridis')
            plt.title("Średnia liczba dni z przekroczeniem normy PM2.5 (15 µg/m³) w województwach")
            plt.xticks(rotation=45, ha='right')
            plt.ylabel("Średnia liczba dni")
            plt.legend(title='Rok')
            plt.tight_layout()
            plt.show()
            
    except Exception as e:
        print(f"Wystąpił błąd: {e}")
```
